```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
url = "https://raw.githubusercontent.com/rolandmueller/titanic/main/titanic3.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df["fare"] = df["fare"].fillna(df["fare"].median())
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

df = df.drop(columns=["cabin"])

df.isnull().sum()
```

```python
df["title"] = df["name"].str.extract(r',\s*([^.]+)\.')

df["title"] = df["title"].replace(
    ["Capt","Col","Major","Dr","Rev","Don","Sir","Lady",
     "Jonkheer","Dona","Countess","Mme","Ms","Mlle"],
    "Rare"
)
```

```python
df["familysize"] = df["sibsp"] + df["parch"] + 1
```

```python
le = LabelEncoder()

df["sex"] = le.fit_transform(df["sex"])
df["title"] = le.fit_transform(df["title"])
```

```python
df = pd.get_dummies(df, columns=["embarked", "pclass"])
```

```python
scaler = StandardScaler()

df[["age", "fare", "sibsp", "parch", "familysize"]] = scaler.fit_transform(
    df[["age", "fare", "sibsp", "parch", "familysize"]]
)
```

```python
df = df.drop(columns=["name", "ticket", "boat", "body", "home.dest"])
```
